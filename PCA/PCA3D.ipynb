```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
s=np.random.seed(2343243)
#method seed() sets the integer starting value used in generating random numbers
#In this way the data has same values every time the random function is called
#For further info:http://www.statisticshowto.com/random-seed-definition/
```

```python
mean_vec1=np.array([0,0,0])
cov_mat1=np.array([[1,0,0],[0,1,0],[0,0,1]])
class1=np.random.multivariate_normal(mean_vec1,cov_mat1,100)
#We are creating 100 data points of class1 with mean as 0
#Draws random samples from a multivariate normal distribution using the mean and covariance of the data
```

```python
mean_vec2=np.array([1,1,1])
cov_mat2=np.array([[1,0,0],[0,1,0],[0,0,1]])
class2=np.random.multivariate_normal(mean_vec2,cov_mat2,100)
#Similarly 100 data points are created for class 2
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D,proj3d
fig=plt.figure(figsize=(8,8))
#This means I need a 8x8 inches figure
ax=fig.add_subplot(111,projection='3d')
#These are subplot grid parameters encoded as a single integer. For example, "111" means "1x1 grid, first subplot" and "234" means "2x3 grid, 4th subplot".
#https://python4mpia.github.io/plotting/advanced.html
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
#class1 data will be marked as a circle
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
plt.show()
```

```python
all_data = np.concatenate((class1,class2))
```

```python
pca=PCA(n_components=2)
transformed_data=pca.fit_transform(all_data)
```

```python
plt.plot(transformed_data[0:100,0],transformed_data[0:100,1],'o')
plt.plot(transformed_data[100:200,0],transformed_data[100:200,1],'^')
plt.show()
```

```python
#Converting data back to 3D
X_approx=pca.inverse_transform(transformed_data)
X_approx
#The data when transformed back may not be same but is along a plane
```
