```python
def step_gradient(points,lr,m,c):
    m_slope=0
    c_slope=0
    for i in range(len(points)):
        x=points[i,0]
        y=points[i,1]
        N=len(points)
        m_slope=m_slope+(-2/N)*(y-m*x-c)*x
        c_slope=c_slope-(-2/N)*(y-m*x-c)
    new_m=m-lr*m_slope
    new_c=c-lr*c
    return new_m,new_c
```

```python
def gd(points,lr,num_iterations):
    m=0
    c=0
    for i in range(num_iterations):
        m,c=step_gradient(points,lr,m,c)
    return m,c
```

```python
import numpy as np
import pandas as pd
data=np.genfromtxt("train_boston.csv",delimiter=",")
finals=[]
for i in range(np.shape(data)[1]):
    for j in range(i+1,np.shape(data)[1]-1):
        x=data[i]
        y=data[j]
        data1=np.array([x,y])
        lr=0.001
        for k in range(10,1000,10):
            num_iterations=k
            m,c=gd(data1,lr,num_iterations)
            finals.append([m,c])
print(finals)
np.savetxt('StepGredfnl.csv',finals,delimiter=",",fmt='%0.5f')
```
